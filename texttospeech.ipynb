```python
from gtts import gTTS

text = "Hello, welcome to Google Colab text to speech demonstration. artificial intelligence is the key technology for todays world."

tts = gTTS(text=text, lang='en', slow=False)

tts.save("output.mp3")

print("Audio saved successfully!")
```
